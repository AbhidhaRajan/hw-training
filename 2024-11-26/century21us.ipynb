```python
import re
from urllib.parse import urlparse
```

```python
url_column = 'profile_url'
urls = df[url_column]
```

```python
missing_urls = df[url_column].isnull().sum()
print(f"Missing URLs: {missing_urls}")
```

```python
def is_valid_url(url):
    regex = re.compile(
        r'^(https?|ftp)://'  # URL must start with http or https
        r'(([A-Z0-9.-]+)+(:[A-Z0-9]*)?)(:[0-9]{1,5})?'
        r'(/?|[/?]\S+)$', re.IGNORECASE)
    return re.match(regex, url) is not None

df['valid_url'] = df[url_column].apply(lambda x: is_valid_url(x) if pd.notnull(x) else False)
```

```python
df['valid_url']
```

```python
invalid_urls = df[~df['valid_url']][url_column]
print("\nInvalid URLs:")
print(invalid_urls)
```

```python
duplicate_urls = df[url_column].duplicated().sum()
print(f"\nDuplicate URLs: {duplicate_urls}")
```
